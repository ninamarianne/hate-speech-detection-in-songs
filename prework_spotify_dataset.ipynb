## Herausfiltern der englischsprachigen Songs 

Erklärung "englishfilter": die Funktion geht alle Elemente der Spalte "language" durch, wenn das Element nicht "en", also die Sprache nicht englisch ist, dann wird der Index dieses Elements in eine neue Liste ("lang_index") geschrieben

```python
def englishfilter():
    lang_index = []
    for lang in enumerate(spotify_dataset.language):
        if lang[1] != "en":
            lang_index.append(lang[0])
    return(lang_index)
```

Erklärung "index_not_english": Liste, die die Indexe aller nicht englischsprachigen Songs enthält

```python
index_not_english = englishfilter()
```

Erklärung "get_index_content": die Funktion nimmt als Argument eine Spalte aus dem Dataset entgegen und gibt eine Liste mit Tupeln zurück, in denen ein Index und der dazugehörige Inhalt aus der Spalte stehen

```python
def get_index_content(column):
    index_content = []
    for content in enumerate(column):
        index_content.append(content)
    return(index_content)
```

```python
#group_information_index
ind_track_name = get_index_content(spotify_dataset.track_name) #Liste an Tupeln aus Index und track name
ind_track_artist = get_index_content(spotify_dataset.track_artist) #Liste an Tupeln aus Index und artist
ind_lyrics = get_index_content(spotify_dataset.lyrics) #Liste an Tupeln aus Index und lyrics
ind_genre = get_index_content(spotify_dataset.playlist_genre) #Liste an Tupeln aus Index und genre
ind_tempo = get_index_content(spotify_dataset.tempo) #Liste an Tupeln aus Index und tempo
ind_energy = get_index_content(spotify_dataset.energy) #Liste an Tupeln aus Index und energy
ind_loudness = get_index_content(spotify_dataset.loudness) #Liste an Tupeln aus Index und loudness
```

Erklärung "get_english_index_content": die Funktion nimmt als Argument eine Liste mit Tupeln aus Index und Inhalt entgegen, sie geht alle Tupel in der Liste durch und prüft, ob der Index in der Liste mit den Indexen der nicht englischsprachigen Songs ist, wenn der Index nicht in der Liste auftaucht, dann soll der dazugehörige Inhalt in eine neue Liste geschrieben werden, d.h. die Funktion gibt eine Liste an den Elementen zurück, die als englischsprachig gelabelt sind

```python
def get_english_index_content(group_information):
    english_index_content = []
    for index, name in group_information:
        if index not in index_not_english:
            english_index_content.append(name)
    return(english_index_content)
```

```python
#group_information_en
en_track_name = get_english_index_content(ind_track_name) #Liste an track names
en_track_artist = get_english_index_content(ind_track_artist) #Liste an artists
en_lyrics = get_english_index_content(ind_lyrics) #Liste an englischsprachigen lyrics
en_genre = get_english_index_content(ind_genre) #Liste an genres
en_tempo = get_english_index_content(ind_tempo) #Liste an tempo numbers
en_energy = get_english_index_content(ind_energy) #Liste an energy numbers
en_loudness = get_english_index_content(ind_loudness) #Liste an loudness numbers
```

Erklärung "makelist": die Funktion nimmt als Argument eine Spalte entgegen und schreibt alle Elemente dieser Spalte in eine Liste

```python
def makelist(dict_words):
    hateful_words = []
    for word in dict_words:
        hateful_words.append(word)
    return (hateful_words)
```

```python
hateful_ngrams = makelist(hate_speech_dict.ngram)
```

## Erstellen einer Liste mit hateful-Lyrics 

Erklärung "make_lyrics_list": die Funktion geht alle Songtexte in dem englischsprachigen Dataset durch und schreibt diese mitsamt ihres Index in ein Tupel in eine Liste

```python
def make_lyrics_list():
    lyrics_list = []
    for lyrics in enumerate (english_spotify_dataset.en_lyrics):
        lyrics_list.append(lyrics)
    return(lyrics_list)
```

```python
index_lyrics_list = make_lyrics_list()
```

Erklärung: "hate_ngram_filter": die Funktion geht alle N-Gramme in der Liste durch und prüft, ob die N-Gramme in den Tupeln der Liste mit den Index-Lyrics-Tupeln vorkommen, wenn das so ist, dann wird der Index des Lyrics in eine Liste geschrieben

```python
def hate_ngram_filter():
    lyrics_index = []
    for ngram in hateful_ngrams:
        for index, lyrics in index_lyrics_list:
            if ngram in lyrics:
                lyrics_index.append(index)
    return(lyrics_index)
```

```python
hateful_song_index = hate_ngram_filter()
```

```python
hateful_song_index.sort() #Sortierung der Größe nach
```

## Entfernen der mehrfachen Indexe 

Da in einigen Songtexten mehrere N-Gramme vorkommen, stehen einige Indexzahlen mehrfach in der Liste lyrics_index.
-> Erklärung "clear_list": die Funktion geht alle Indexzahlen durch und schreibt diejenigen in eine neue Liste, die dort noch nicht drin stehen, d.h. sie sorgt dafür, dass alle Indexzahlen nur einmal vorkommen

```python
def clear_list():
    cleared_hateful_song_index = []

    for index in hateful_song_index:
        if index not in cleared_hateful_song_index:
            cleared_hateful_song_index.append(index)

    return(cleared_hateful_song_index)
```

```python
cleared_lyrics_index = clear_list()
```
