```python
import pandas as pd
```

## Einlesen der Songdatei und der Datei mit den Labels 

```python
labels = pd.read_csv("data/labeled_songs.csv")
hate_spotify_dataset = pd.read_csv("data/hate_spotify_songs.csv")
```

```python
labels #0 = Hate Speech, 2 = offensive language, 2 = neither
```

```python
hate_spotify_dataset
```

## Zusammenführen der beiden Dateien 

```python
labeled_dataset = pd.DataFrame({'track_name': hate_spotify_dataset.track_name,
                    'track_artist': hate_spotify_dataset.track_artist,
                    'lyrics': hate_spotify_dataset.lyrics,
                    'genre': hate_spotify_dataset.genre,
                    'tempo': hate_spotify_dataset.tempo,
                    'energy': hate_spotify_dataset.energy,
                    'loudness': hate_spotify_dataset.loudness,
                    'label': labels.label})

labeled_dataset_df = pd.DataFrame(labeled_dataset)

labeled_dataset_df.to_csv("data/labeled_dataset.csv")
```
